```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import FactorRange
import pandas as pd
```

```python
output_notebook()
```

```python
file = "../data/income_by_age.tsv"
df = pd.read_table(file)

race_data = (
    df.sort_values(["race", "age"])
    .reset_index(drop=True)
    .iloc[7:35, :]
    .reset_index(drop=True)
)
age_group = race_data.groupby(["age", "race"])[["median_income"]].sum()
```

```python
age_group
```

```python
from bokeh.models import NumeralTickFormatter as NTF
from bokeh.palettes import Blues8

colors = Blues8[4]

p = figure(
    title="Median income by age",
    height=300,
    sizing_mode="stretch_width",
    # width=500,
    x_range=FactorRange(factors=age_group.index),
)

p.vbar(
    x=age_group.index,
    top=age_group.median_income,
    width=0.9,
    # legend_label=age_group.index)
)

p.y_range.start = 0
p.y_range.end = 100_000
p.yaxis.formatter = NTF(format="$0,0")

# remove x-axis tick labels and tick marks
p.xaxis.major_label_text_font_size = "0pt"
p.xaxis.major_label_text_font_style = "normal"
p.xaxis.major_tick_line_color = None
p.xaxis.major_tick_out = 0

show(p)
```
